```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Import pandas for DataFrame manipulation

# Establecer el tamaÃ±o de las figuras usando seaborn
sns.set(rc={'figure.figsize':(11, 4)})

# Load the dataset (replace with your actual file path or URL)
df = pd.read_csv('https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv')

# Filter the DataFrame for 'Garfield-South Elevated' station
Garfield_South_Elevated = df[df['stationname'] == 'Garfield-South Elevated']

# Convert 'date' column to datetime and set it as index
Garfield_South_Elevated['date'] = pd.to_datetime(Garfield_South_Elevated['date'])
Garfield_South_Elevated.set_index('date', inplace=True)

# Now you can plot the data
plt.plot(Garfield_South_Elevated.index, Garfield_South_Elevated['rides'], linewidth=0.5)

# PersonalizaciÃ³n del grÃ¡fico
plt.title('NÃºmero de Viajes en Garfield-South Elevated')  # TÃ­tulo del grÃ¡fico
plt.xlabel('Fecha')  # Etiqueta del eje x
plt.ylabel('NÃºmero de Viajes')  # Etiqueta del eje y
plt.grid(True)  # Habilitar cuadrÃ­cula en el grÃ¡fico
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor visualizaciÃ³n

# Mostrar el grÃ¡fico
plt.tight_layout()  # Ajustar diseÃ±o para evitar superposiciÃ³n de elementos
plt.show()
```
